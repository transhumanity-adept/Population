```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
import os
```

```python
from pathlib import Path

def root_path() -> Path:
    return Path(os.path.abspath('')).parent.parent.parent 
```

```python
PATH_DATA = os.path.join(root_path(), 'data\\regions')
PATH_DATA
```

```python
FILE_NAME_CVS: list[str] = os.listdir(PATH_DATA)
FILE_NAME_PATH_CVS = list(map(lambda file_name: f"{PATH_DATA}\{file_name}", FILE_NAME_CVS))
FILE_NAME_PATH_CVS
```

```python
#Библиотечная экстраполяция - готово
data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")
x = np.array(data.columns)
x = list(map(int, x))
xnew = np.arange(2023, 2027)
tmp = len(data.values[0])
for i in range(0, len(data.index)):
    y = np.array(data.values[i, 0:tmp])
    f = interpolate.interp1d(x, y, fill_value='extrapolate')
    ynew = f(xnew)
    for t in range(0, len(xnew)):
        if (ynew[t] < 0):
            ynew[t] = 0
        data.at[i, xnew[t]] = ynew[t]
data
```

```python
# Проверка точности метода выше на известных данных
data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")
x = np.array(data.columns)
x = list(map(int, x))
for n in range(0, 3):
    x = np.delete(x, 10, 0)
xnew = np.arange(2020, 2023)
tmp = len(x)
for i in range(0, len(data.index)):
    ytot = np.array(data.values[i])
    y = np.array(data.values[i, 0:tmp])
    f = interpolate.interp1d(x, y, fill_value='extrapolate')
    ynew = y
    ynew2 = f(xnew)
    ynew = np.append(ynew, ynew2)
    print((1-ytot/ynew)*100)
```

```python
# МНК - в процессе
data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")
x = np.array(data.columns)
x = np.array(list(map(int, x)))
x = x-2009
xnew = np.arange(2023, 2027)-2009
tmp = len(data.values[0])
y = np.array(data.values[0])
n = 5
matrixX = [[0]*n for i in range(len(x))]
for i in range(0, len(x)):
    for j in range(0, n):
        matrixX[i][j] = x[i]**j
y = np.array(data.values[0])
matrixX = np.array(matrixX)
print(matrixX)
A = y.dot(matrixX)
print(A)
ynew = A[0]
for a in range(1, n):
    ynew += A[a]*(14**a)
print(ynew)
```
