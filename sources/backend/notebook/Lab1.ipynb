```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
from sklearn.linear_model import LinearRegression
import os
```

```python
from pathlib import Path

def root_path() -> Path:
    return Path(os.path.abspath('')).parent.parent.parent 
```

```python
PATH_DATA = os.path.join(root_path(), 'data\\regions')
PATH_DATA
```

```python
FILE_NAME_CVS: list[str] = os.listdir(PATH_DATA)
FILE_NAME_PATH_CVS = list(map(lambda file_name: f"{PATH_DATA}\{file_name}", FILE_NAME_CVS))
FILE_NAME_PATH_CVS
```

```python
#Библиотечная экстраполяция - готово
data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")
x = np.array(data.columns)
x = list(map(int, x))
xnew = np.arange(2023, 2027)
tmp = len(data.values[0])
for i in range(0, len(data.index)):
    y = np.array(data.values[i, 0:tmp])
    f = interpolate.interp1d(x, y, fill_value='extrapolate')
    ynew = f(xnew)
    for t in range(0, len(xnew)):
        if (ynew[t] < 0):
            ynew[t] = 0
        data.at[i, xnew[t]] = ynew[t]
data
```

```python
# Проверка точности метода выше на известных данных
data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")
x = np.array(data.columns)
x = list(map(int, x))
for n in range(0, 3):
    x = np.delete(x, 10, 0)
xnew = np.arange(2020, 2023)
tmp = len(x)
for i in range(0, len(data.index)):
    ytot = np.array(data.values[i])
    y = np.array(data.values[i, 0:tmp])
    f = interpolate.interp1d(x, y, fill_value='extrapolate')
    ynew = y
    ynew2 = f(xnew)
    ynew = np.append(ynew, ynew2)
    print((1-ytot/ynew)*100)
```

```python
# МНК - в процессе
data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")
x = np.array(data.columns)
x = np.array(list(map(int, x)))
x = x-2009
xnew = np.arange(2023, 2027)-2009
tmp = len(data.values[0])
y = np.array(data.values[0])
y = y-np.mean(y)
print(y)
n = len(data.values[0])
matrixX = [[0]*n for i in range(len(x))]
for i in range(0, len(x)):
    for j in range(0, n):
        matrixX[i][j] = x[i]**j
y = np.array(data.values[0])
matrixX = np.array(matrixX)
matrixX = np.linalg.inv(matrixX)
# print(matrixX)
A = matrixX.dot(y)
print(A)
ynew = A[0]
for a in range(1, n):
    ynew += A[a]*(14**a)
print(ynew)
```

```python
# Линейная регрессия
data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")

# с 2009 по 2019
null_yaer = data.iloc[:1, :10] 


X = np.array(list(map(lambda year: int(year), null_yaer.columns))).reshape((-1, 1))
y = null_yaer.values[0]



X_predict = np.array([2020, 2021, 2022]).reshape((-1, 1))
y_predict = LinearRegression().fit(X, y).predict(X_predict)


for k,v in dict(zip([2020, 2021, 2022], y_predict)).items():
    null_yaer[k] = round(v, 0)

# print(data.iloc[:1, :])
null_yaer


# null_yaer = pd.DataFrame(dict(zip([2020, 2021, 2022], y_predict)), index=['0'])
```

```python
# полиномиально
from sklearn.preprocessing import PolynomialFeatures

data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")

null_yaer = data.iloc[:1, :10]


X = np.array(list(map(lambda year: int(year), null_yaer.columns))).reshape((-1, 1))
y = null_yaer.values[0]

polinom=PolynomialFeatures() 
X_pol = polinom.fit_transform(X)    

X_predict = np.array([2020, 2021, 2022]).reshape((-1, 1))
y_predict = LinearRegression().fit(X_pol, y).predict(polinom.fit_transform(X_predict))


for k,v in dict(zip([2020, 2021, 2022], y_predict)).items():
    null_yaer[k] = round(v, 0)

data_check = data.iloc[:1, :]
```

```python
from bokeh.plotting import figure, show
p=figure(title="Предсказание численнсоти населения полиномиально", width=800, height=800)
p.title.align = 'center'
p.circle(data.columns, data.values[0])
p.line(data.columns, data.values[0], legend_label='Статистическая информация', line_width=3, line_alpha=0.4)
p.circle(null_yaer.columns, null_yaer.values[0])
p.line(null_yaer.columns,null_yaer.values[0], color="red", legend_label='Предсказание с 2020 по 2022', line_width=3, line_alpha=0.4)
p.xaxis.axis_label = 'Год'
p.yaxis.axis_label = 'Численность'
show(p)
```

```python
# полиномиально до 2100
from sklearn.preprocessing import PolynomialFeatures

data = pd.read_csv(PATH_DATA + '\Брянская область.csv',
                   index_col="Age", sep=";")

null_yaer = data.iloc[:1]
null_yaer

X = np.array(list(map(lambda year: int(year), null_yaer.columns))).reshape((-1, 1))
y = null_yaer.values[0]

polinom=PolynomialFeatures() 
X_pol = polinom.fit_transform(X)    

X_predict = np.array(range(2023, 2100)).reshape((-1, 1))
y_predict = LinearRegression().fit(X_pol, y).predict(polinom.fit_transform(X_predict))


for k,v in dict(zip(range(2023, 2101), y_predict)).items():
    null_yaer[k] = round(v, 0)

data_check = data.iloc[:1, :]
```

```python
from bokeh.plotting import figure, show, output_file
p=figure(title="Предсказание численнсоти населения полиномиально", width=800, height=800)
p.title.align = 'center'
p.circle(data_check.columns, data_check.values[0])
p.line(data_check.columns, data_check.values[0], legend_label='Статистическая информация', line_width=3, line_alpha=0.4)
p.circle(null_yaer.columns, null_yaer.values[0])
p.line(null_yaer.columns,null_yaer.values[0], color="red", legend_label='Предсказание с 2023 по 2100', line_width=3, line_alpha=0.4)
p.xaxis.axis_label = 'Год'
p.yaxis.axis_label = 'Численность'
show(p)
```
